```
from pathlib import Path
```

```
src = Path('/content/drive/My Drive/QuodAvis/SpotTheBird!/data')
dest = Path('/content/drive/My Drive/QuodAvis/SpotTheBird!/path')
src.exists() and dest.exists()
```

```
for x in src.iterdir():
    p = dest / 'train' / x.name
    p.mkdir(exist_ok=True)
    p = dest / 'valid' / x.name
    p.mkdir(exist_ok=True)
    p = dest / 'test' / x.name
    p.mkdir(exist_ok=True)
```

```
for x in src.iterdir():
    count = 0
    for y in x.iterdir():
        if count < 40:
            destination = dest / 'train' / x.name / y.name
            y.replace(destination)
        elif count < 50:
            destination = dest / 'valid' / x.name / y.name
            y.replace(destination)
        else:
            destination = dest / 'test' / x.name / y.name
            y.replace(destination)
        count = count + 1
```

```
count = 0
for x in src.iterdir():
    try:
        os.rmdir(x)
    except OSError as ex:
        if ex.errno == errno.ENOTEMPTY:
            count = count + 1

print('The number of non-empty dirs is ' + str(count) + '.')
```
